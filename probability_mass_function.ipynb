```python
import matplotlib.pyplot as plt

# Possible outcomes of a die roll
outcomes = [1, 2, 3, 4, 5, 6]
probabilities = [1/6] * 6  # Equal probability for each outcome

# Plotting the PMF
plt.bar(outcomes, probabilities, width=0.6, color='skyblue', edgecolor='black')
plt.title("PMF of a Fair Die Roll")
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.show()
```
