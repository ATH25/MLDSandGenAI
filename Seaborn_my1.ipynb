# Matplotlib Example:

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(0, 10, 100)
# print (x)
y = np.sin(x)

# Create a line plot
plt.plot(x, y, label="sin(x)", color="blue")
plt.title("Matplotlib Example")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.show()
```
