# Standard Deviation and Variance

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

incomes = np.random.normal(100.0, 50.0, 10000)

plt.hist(incomes, 50, color='green', edgecolor='red')

plt.title("Income Distribution")
plt.xlabel("Income")
plt.ylabel("Frequency")

plt.show()
```

```python
incomes.std()
```

```python
incomes.size
```

```python
incomes.var()
```

```python
i = 1
for income in incomes:
    print(f"{i}: {income}")
    i += 1
```
